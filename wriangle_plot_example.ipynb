```python
from getdist import plots, gaussian_mixtures
samples = gaussian_mixtures.randomTestMCSamples(ndim=4, nMCSamples=6)
g = plots.get_subplot_plotter()
```

```python
# classical triangle
g.triangle_plot(samples, filled=True, legend_labels = ["sample %d"%i for i in range(len(samples))])
                                                    
```

```python
# whiskers !
g.triangle_plot(samples, filled=True, legend_labels = ["sample %d"%i for i in range(len(samples))],diag1d_kwargs={"whisker":True})
```

```python
# with a reference line for sample 4
g.triangle_plot(samples, filled=True, legend_labels = ["sample %d"%i for i in range(len(samples))],
                diag1d_kwargs={"whisker":True,"whisker_ref":4})
```

```python
# lets change the color, line style and thickness of the ref line 
g.triangle_plot(samples, filled=True, legend_labels = ["sample %d"%i for i in range(len(samples))],
                diag1d_kwargs={"whisker":True,
                               "whisker_ref":4,"whisker_ref_lw":2,"whisker_ref_color":"orange","whisker_ref_ls":":"})
```

```python
# lets change the style of the whiskers
g.triangle_plot(samples, filled=True, legend_labels = ["sample %d"%i for i in range(len(samples))],
                diag1d_kwargs={"whisker":True,
                               "whisker_ref":4,"whisker_ref_lw":2,"whisker_ref_color":"orange","whisker_ref_ls":":",
                               "whisker_capsize":3,"whisker_mew":[1,2,3,4,5,6],"whisker_marker":["o","d","x","v","^","D"],"lws":2})
```

```python
# # let's add the distributions as well
g.triangle_plot(samples, filled=True, legend_labels = ["sample %d"%i for i in range(len(samples))],
                diag1d_kwargs={"whisker":True,
                               "whisker_ref":4,"whisker_ref_lw":2,"whisker_ref_color":"orange","whisker_ref_ls":":",
                               "whisker_capsize":3,"whisker_mew":[1,2,3,4,5,6],"whisker_marker":["o","d","x","v","^","D"],
                              "whisker_both":True})
```

```python
# let's add the distributions and change style
g.triangle_plot(samples, filled=True, legend_labels = ["sample %d"%i for i in range(len(samples))],
                diag1d_kwargs={"whisker":True,
                               "whisker_ref":4,"whisker_ref_lw":2,"whisker_ref_color":"orange","whisker_ref_ls":":",
                               "whisker_capsize":7,"whisker_mew":[1,2,3,4,5,6],"whisker_marker":["o","d","x","v","^","D"],
                              "whisker_both":True, "whisker_both_color":"grey", "whisker_both_cut":4,"whisker_both_ls":"--","whisker_both_lw":3})
```

```python
# add mean and errors
g.triangle_plot(samples, filled=True, legend_labels = ["sample %d"%i for i in range(len(samples))],
                diag1d_kwargs={"whisker":True,
                               "whisker_ref":0,"whisker_ref_lw":.5,"whisker_ref_color":"k","whisker_ref_alpha":.5,
                               "whisker_capsize":0,"whisker_mew":2,"whisker_marker":["o","d","x","v","^","D"],
                               "whisker_print_mean":True})
```

```python
# add mean and errors with marginal
g.triangle_plot(samples, filled=True, legend_labels = ["sample %d"%i for i in range(len(samples))],
                diag1d_kwargs={"whisker":True,
                               "whisker_ref":0,"whisker_ref_lw":.5,"whisker_ref_color":"k","whisker_ref_alpha":.5,
                               "whisker_capsize":0,"whisker_mew":2,"whisker_marker":["o","d","x","v","^","D"],
                               "whisker_print_mean":True, "whisker_both":True,"whisker_both_alpha":.3})
```

```python
# add mean and error
# Do not print the second and last value 
# The first value has a special color (pink) the other one have the same color as the whisker
# Default padding from the whisker is .2. Add more space for the first and 4th, 
# and give a negative padding for the third value so that it appears below the whisker, and set its alpha to .5
# pick a larger font size for the thrid value
# default is to merge the two limits if their difference is < mean(top+bottom)/10
# change the precision for the 5th value to mean(top+bottom)/100, and use 5 significative digits

g.triangle_plot(samples, filled=True, legend_labels = ["sample %d"%i for i in range(len(samples))],
                diag1d_kwargs={"whisker":True,
                               "whisker_ref":0,"whisker_ref_lw":.5,"whisker_ref_color":"k","whisker_ref_alpha":.5,
                               "whisker_capsize":0,"whisker_mew":2,"whisker_marker":["o","d","x","v","^","D"],
                               "whisker_print_mean":[True,False,True,True,True,False],"whisker_print_mean_color":["pink"],
                               "whisker_print_mean_pad":[.3,None,-.6,.3],
                               "whisker_print_mean_fontsize":[None,None,10],
                               "whisker_print_mean_precision":[None, None,None,None,100],
                               "whisker_print_mean_ndigits":[None, None,None,None,5],
                               "whisker_print_mean_alpha":[None,None,.5]})
```
